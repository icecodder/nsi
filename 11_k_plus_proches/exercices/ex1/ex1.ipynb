2.	Compléter le  programme pour afficher dans la console le retour de distances(listeX, listeY, mystere).


```python
def distance_E1(xA, yA, xB, yB):
    return sqrt((xB - xA) ** 2 + (yB - yA) ** 2)


def distance_E2(a, b):
    return sqrt((b[0] - a[0]) ** 2 + (b[1] - a[1]) ** 2)


def distances(listeX, listeY, mystere):
    #mystere[0] : abscisse
    #mystere[1] : ordonnées
    #Retourne la liste des distances entre mystere et les autres éléments
    
    liste_distances = []
    
    for i in range(len(listeX)):
        d = distance_E1(listeX[i], listeY[i], mystere[0], mystere[1])
        liste_distances.append(d)
        
    return liste_distances
```

```python
def choix_premier(element):
    return element[0]


def k_plus_proches_voisins(listeX, listeY, mystere, k):
    #Retourne la liste des k plus proches voisins
    #Retourne une liste de couples 

    # Calculer les distances
    liste_distances = distances(listeX, listeY, mystere)
    
    # Associer distance et famille
    liste_distances_familles = list(zip(liste_distances, liste_famille)) # Transformer deux listes en une liste de couples
     
     
    # Trier la liste en fonction de la distance
    distance_triee = sorted(liste_distances_familles, key=choix_premier)
    
    # Choisir les k premiers
    #k_premiers = [distance_triee[i] for i in range(k)]
    k_premiers = distance_triee[:k]
      
    return k_premiers
```
