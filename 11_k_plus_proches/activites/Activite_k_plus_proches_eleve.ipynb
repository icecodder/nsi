### 2.2 Représentation graphique


```python
import matplotlib.pyplot as plt


famille1_vie     = [50, 90, 75]
famille1_attaque = [50, 80, 90]

famille2_vie     = [50, 80, 90, 55]
famille2_attaque = [95, 45, 75, 80]

carte_mystere_vie = [65]
carte_mystere_attaque = [70]

plt.figure()

plt.scatter(famille1_vie, famille1_attaque, color='g', label='Famille 1')
plt.scatter(famille2_vie, famille2_attaque, color='r', label='Famille 2')
plt.scatter(carte_mystere_vie, carte_mystere_attaque, color='b', label='Carte mystère')

plt.xlabel("Points de vie")
plt.ylabel("Points d'attaque")

plt.legend()
plt.grid()

plt.show()
```

### 2.3 Algorithme des k plus proches

```python
from math import sqrt


def distance(a, b) :
    return sqrt((b[0] - a[0]) ** 2 + (b[1] - a[1]) ** 2)


m = (65, 70)
p1 = (50, 50)
p2 = (90, 80)
p3 = (75, 90)
p4 = (50, 95)
p5 = (80, 45)
p6 = (90, 75)
p7 = (55, 80)

print("p1 :", distance(m, p1))
print("p2 :", distance(m, p2))
print("p3 :", distance(m, p3))
print("p4 :", distance(m, p4))
print("p5 :", distance(m, p5))
print("p6 :", distance(m, p6))
print("p7 :", distance(m, p7))
```
